Univariate Analysis

```python
plt.rcParams['figure.figsize'] = [40, 40]
fig,ax = plt.subplots(3,2)
ax[0,0].hist(df['number of bedrooms'])
ax[0,1].pie(df['number of bathrooms'].value_counts(), labels=df['number of bathrooms'].value_counts().index)
ax[1,0].boxplot(df['living area'])
ax[1,1].boxplot(df['lot area'])
ax[2,0].pie(df['number of floors'].value_counts(), labels=df['number of floors'].value_counts().index)
ax[2,1].hist(df['Built Year'])
plt.show()
```
